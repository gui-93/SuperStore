```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados
sample = pd.read_csv('sample-superstore.csv')
```

```python
print(sample.describe())
```

```python
print(sample.head())
```

```python
print(sample.isnull().sum())
```

```python
print(sample.dtypes)
```

```python
# Formatar as colunas
sample['Order Date'] = pd.to_datetime(sample['Order Date'])
sample['Ship Date'] = pd.to_datetime(sample['Ship Date'])
sample['Sales'] = sample['Sales'].str.replace(',', '.').astype(float)
sample['Quantity'] = sample['Quantity'].astype(int)
sample['Discount'] = sample['Discount'].str.replace(',', '.').astype(float)
sample['Profit'] = sample['Profit'].str.replace(',', '.').astype(float)
sample['Postal Code'] = sample['Postal Code'].astype(str)
```

```python
# Renomear o nome das colunas
sample = sample.rename(columns={
    'Row ID': 'ID da Linha',
    'Order ID': 'ID do Pedido',
    'Order Date': 'Data do Pedido',
    'Ship Date': 'Data de Envio',
    'Ship Mode': 'Modo de Envio',
    'Customer ID': 'ID do Cliente',
    'Customer Name': 'Nome do Cliente',
    'Segment': 'Segmento',
    'Country': 'País',
    'City': 'Cidade',
    'State': 'Estado',
    'Postal Code': 'Código Postal',
    'Region': 'Região',
    'Product ID': 'ID do Produto',
    'Category': 'Categoria',
    'Sub-Category': 'Subcategoria',
    'Product Name': 'Nome do Produto',
    'Sales': 'Vendas',
    'Quantity': 'Quantidade',
    'Discount': 'Desconto',
    'Profit': 'Lucro'
})
```

```python
print(sample.columns)
```

```python
print(sample.head())
```

# Análises Exploratórias

```python
# Análise de Vendas por Categoria
sample.groupby('Categoria')['Vendas'].sum().plot(kind='bar')
plt.title('Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Vendas')
plt.show()
```

#### O gráfico mostra que a categoria "Technology" tem as maiores vendas, seguida pela categoria "Furniture".

___________________________________________________________________________________________________________________________________________________________

```python
# Análise de Lucro por Categoria
sample.groupby('Categoria')['Lucro'].sum().plot(kind='bar')
plt.title('Lucro por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Lucro')
plt.show()
```

#### O gráfico mostra que a categoria "Technology" tem o maior lucro, seguida pela categoria "Office Supplies". Aqui mostra que, mesmo "Furniture" vendendo mais que "Office Supplies" (no gráfico de Vendas), no final "Office Supplies" gera mais lucro que "Furniture".

___________________________________________________________________________________________________________________________________________________________

```python
#Análise de Desconto por Categoria
sample.groupby('Categoria')['Desconto'].mean().plot(kind='bar')
plt.title('Desconto por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Desconto')
plt.show()
```

#### O gráfico revela uma tendência interessante: a categoria "Furniture" oferece os maiores descontos. Isso pode ser um dos principais motivos que contribui para o menor lucro da categoria.Por outro lado, a categoria "Technology" apresenta um cenário oposto, oferecendo menores descontos, essa categoria consegue manter uma margem de lucro mais saudável, o que pode ser um dos fatores que a torna a mais lucrativa.

___________________________________________________________________________________________________________________________________________________________

```python
# Análise de Vendas por Região
sample.groupby('Região')['Vendas'].sum().plot(kind='bar')
plt.title('Vendas por Região')
plt.xlabel('Região')
plt.ylabel('Vendas')
plt.show()
```

#### O gráfico mostra que a região "West" tem as maiores vendas, seguida pela região "South".

___________________________________________________________________________________________________________________________________________________________

```python
# Análise de Vendas por Mês
sample['Data do Pedido'] = pd.to_datetime(sample['Data do Pedido'])
sample['Mês'] = sample['Data do Pedido'].dt.month
sample.groupby('Mês')['Vendas'].sum().plot(kind='bar')
plt.title('Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.show()
```

#### O gráfico mostra que as vendas são mais altas nos meses de novembro e dezembro. O que pode ser devido à demanda por produtos durante a temporada de festas.
